# 2. 객체 분류

## Challenges
- Illumination
    - 조명의 조건에 따라 객체의 정보가 달라짐
- Deformation
    - 물체의 형태, 포즈, 크기등이 변화함
- Occlusion
    - 다른 물체에 의해 부분적으로 가려짐
    - 부분 가림: 물체의 일부분이 다른 물체에 의해 가려짐
    - 자체 가림: 물체의 한 부분이 동일 물체의 다른 부분에 의해 가려짐
    - 배경 혼합: 물체와 배경의 경계가 명확하지 않아 발생하는 가림 효과
    - 정보 손실: 가려진 부분의 특징 정보가 손실되어 인식 정확도 저하
- background clutter
    - 객체가 찍힌 사진의 배경과 유사하여 분류하기 어려워 질수있음
- intraclass variation
    - 클래스 내의 객체 간에도 사소한 차이가 있음

# 3. 손실함수

## 손실함수
- 주어진 데이터셋에 대해 분류기의 성능을 평가하는데 사용
- $L = \frac{1}{N} \sum\limits_{i=1}^{N} L_i(f(x_i, W), y_i)$
    - 각 분류기에 대한 손실함수 출력의 평균임
- 대표적 손실 함수
    - 분류
        - BCE
            - $L_{BCE} = -\frac{1}{N} \sum\limits_{i=1}^{N} [y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i)]$
        - CCE
            - $L_{CCE} = -\frac{1}{N} \sum\limits_{i=1}^{N} \sum\limits_{j=1}^{M} y_{ij} \log(\hat{y}_{ij})$
    - 회귀
        - MSE
            - $L_{MSE} = \frac{1}{N} \sum\limits_{i=1}^{N} (y_i - \hat{y}_i)^2$
        - MAE
            - $L_{MAE} = \frac{1}{N} \sum\limits_{i=1}^{N} |y_i - \hat{y}_i|$



## 정규화
- 3-2-1 참조
- 출력값 범위 제어
    - 신경망 출력 값의 기하학적 모형 원, 구형으로 변환함
- 아웃라이어에 대한 과적합 방지
- 정규화 종류
    - L1 정규화
        - $R(w) = ||w||_1 = \sum_{i=1}^{n} |w_i|$
    - L2 정규화
        - $R(w) = ||w||_2^2 = \sum_{i=1}^{n} w_i^2$
    - Elastic Net
        - $R(w) = \alpha \sum_{i=1}^{n} |w_i| + (1-\alpha) \sum_{i=1}^{n} w_i^2$
        - L1, L2 결합임
