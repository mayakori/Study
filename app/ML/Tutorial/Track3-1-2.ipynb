- ## 순전파
    - 

    - 
        - 순전파, 역전파 표현
    - $ z^{[l]} = W^{[l]} a^{[l-1]} + b^{[l]} $

        $ a^{[l]} = g^{[l]}(z^{[l]}) $

    - 마지막 레이어 도달할떄까지 해당 연산 반복

    - 훈련 데이터 단위로 벡터화

    - 순전파 과정에서 변화하는 출력 행렬의 크기를 주시할것

        - $ W^{[l]}: (n^{[l]}, n^{[l-1]}) $

            $ b^{[l]}: (n^{[l]}, 1) $
        
        - $l$ 레이어의 가중치, 편향 행렬 크기
        
        - 도함수 또한 같은 차원을 가짐

        - ### 벡터화시

            - $ z^{[l]}, a^{[l]} : (n^{[l]}, 1) $

                $ \rightarrow Z^{[l]}, A^{[l]} : (n^{[l]}, m) $

                $ l = 0 \quad A^{[0]} \cdot X = (n^{[0]}, m) $

                $ dZ^{[l]}, dA^{[l]} : (n^{[l]}, m) $

            - 입력 데이터와 출력 데이터의 크기가 $(n,1)$ 에서 $(n,m)$으로 달라졌음
